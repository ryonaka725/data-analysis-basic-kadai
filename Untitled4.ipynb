```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")

def tax(x):
    return x * 1.10

tax_series = df['単価'].apply(tax)
tax_series.name = "単価(税込み)"
pd.concat([df, tax_series], axis=1)
```

```python
!pip install japanize_matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

df = pd.read_csv('sample_pandas_6.csv')
category_df = pd.read_csv('category.csv')
category_df

# カテゴリーの列をマージ
df= pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df

# カテゴリーの出現頻度を棒グラフ化
df['カテゴリー'].value_counts().plot(kind='bar')
plt.title('カテゴリーの出現頻度')
plt.tight_layout()
plt.show()

# データフレームを基準列ごとに要素をグループにまとめ、サブ列を集計するメソッド
df.groupby('商品番号')['注文数'].sum()

# データフレームの基本統計量を算出するメソッド
df.groupby('商品番号')['注文数'].describe()

```
